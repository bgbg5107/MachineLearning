### 1. svc로 iris 데이터 분류

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC #support vector c
```

### 2. 데이터 수집

```python
df = load_iris()
```

```python
x_data = pd.DataFrame(df.data, columns = df.feature_names)
y_data = pd.DataFrame(df.target, columns = ['species'])
```

```python
x_data
```

```python
y_data
```

### 3. 데이터 전처리
- 스케일링

```python
scaler = MinMaxScaler()
scaler.fit(x_data)
x_data = scaler.transform(x_data)

#scaler.fit_transform(x_data) 바로 변형가능
```

```python
x_data
```

### 레이블 인코딩

```python
le = LabelEncoder()
le.fit(y_data)
print(le.classes_)
```

```python
labels = ['세토사','버시칼라','버지니카']
```

```python
y_data = le.transform(y_data)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=0, stratify=y_data)
```

```python
x_train.shape
```

```python
y_train.shape
```

### 4. 탐색적 분석

### 5-7. 모델링(생성, 학습, 평가)

```python
model = SVC(kernel='rbf', gamma='auto', C=0.1, probability=True)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
