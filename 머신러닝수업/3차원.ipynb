```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df
```

```python
tsne = TSNE(n_components=2).fit_transform(df)
```

```python
tsne
```

```python
tsne_df = pd.DataFrame(tsne, columns=['component0', 'component1'])
```

```python
tsne_df
```

#### t-sne 2차원 데이터 시각화

```python
tsne_df['target'] = iris.target
```

```python
tsne_df_0 = tsne_df[tsne_df['target']==0]
tsne_df_1 = tsne_df[tsne_df['target']==1]
tsne_df_2 = tsne_df[tsne_df['target']==2]
```

```python
plt.scatter(tsne_df_0['component0'], tsne_df_0['component1'], color='pink', label='setosa')
plt.scatter(tsne_df_1['component0'], tsne_df_1['component1'], color='purple', label='versicolor')
plt.scatter(tsne_df_2['component0'], tsne_df_2['component1'], color='yellow', label='virginica')
plt.xlabel('component0')
plt.ylabel('component1')
plt.legend()
plt.show()
```

#### 3차원 t-sne 시각화

```python
# t-sne 모델 불러오기
# 학습 + 변환 
tsne = TSNE(n_components=3).fit_transform(df)
tsne_df = pd.DataFrame(tsne, columns=['component0','component1','component2'])
```

```python
tsne_df
```

```python
tsne_df['target'] = iris.target #target 열추가
tsne_df
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111, projection='3d')

tsne_df_0 = tsne_df[tsne_df['target']==0]
tsne_df_1 = tsne_df[tsne_df['target']==1]
tsne_df_2 = tsne_df[tsne_df['target']==2]

ax.scatter(tsne_df_0['component0'], tsne_df_0['component1'], tsne_df_0['component2'], color='pink', label='setosa')
ax.scatter(tsne_df_1['component0'], tsne_df_1['component1'], tsne_df_1['component2'], color='purple', label='versicolor')
ax.scatter(tsne_df_2['component0'], tsne_df_2['component1'], tsne_df_2['component2'], color='yellow', label='virginica')

ax.set_xlabel('component0')
ax.set_ylabel('component1')
ax.set_zlabel('component2')
ax.legend()
plt.show()
```
