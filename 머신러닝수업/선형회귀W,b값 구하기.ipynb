```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = [0,1,2,3,4]
y = [2,4,6,8,10]
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

### 모델

- 선형회귀 모델 만들기
- 만들어진 모델에서 w,b 값 확인하기

```python
x_train = pd.DataFrame(x)
y_train = pd.DataFrame(y)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict([[6]])
```

```python
print(lr.coef_) #기울기값, w값
print(lr.intercept_) #절편값
```
